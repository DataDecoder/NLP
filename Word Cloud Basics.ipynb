### Default Word Cloud

```python
# !pip install wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# generate the w_cloud
wordcloud = WordCloud().generate(text)

# plot the w_cloud
plt.figure(figsize = (12, 5))
plt.imshow(wordcloud)
plt.show()
```

### Removing words that are not important/required as per the user's choice

```python
# generate the w_cloud
wordcloud = WordCloud(stopwords = ['Flauius', 'Murellus', 'Sir']).generate(text)

# plot the w_cloud
plt.figure(figsize = (12, 5))
plt.imshow(wordcloud)
plt.show()
```

### Changing the background, palette of w_cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# generate the w_cloud
wordcloud = WordCloud(width = 1080, height = 1080, background_color = 'black', colormap = 'Blues').generate(text)

# plot the w_cloud
plt.figure(figsize = (5, 5))
plt.imshow(wordcloud)
plt.show()
```
